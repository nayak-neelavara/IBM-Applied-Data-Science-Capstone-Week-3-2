## 1. Import Libraries

```python
import pandas as pd
import numpy as np
```

## 2. Create neighborhoods dataframe

```python
df_neighborhoods = pd.read_csv('downloads/Toronto_Postcodes_1.csv',index_col=[0])
df_neighborhoods.head(10)
```

## 3. Create coordinates dataframe

```python
df_coordinates = pd.read_csv('downloads/Geospatial_Coordinates.csv')
df_coordinates.head(10)
```

## 4. Rename Postcode column of neighborhoods dataframe

```python
df_neighborhoods.rename(columns={'Postcode': 'Postal Code'}, inplace=True)
df_neighborhoods.head(10)
```

## 5. Merge the two dataframes on column 'Postal Code'

```python
df_neighborhoods_coordinates = pd.merge(df_neighborhoods, df_coordinates, on='Postal Code')
df_neighborhoods_coordinates.head(10)
```

## 6. Export the dataset to .csv file

```python
df_neighborhoods_coordinates.to_csv('downloads/Toronto_Postcodes_2.csv')
```
